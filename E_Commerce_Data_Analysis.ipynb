```python
import pandas as pd
import plotly.express as px
import plotly.graph_objects as go
import plotly.io as pio
import plotly.colors as colors
pio.templates.default="plotly_white"
```

```python
data=pd.read_csv("/content/Sample - Superstore.csv", encoding='latin1')
```

```python
#Converting date column
data['Order Date']=pd.to_datetime(data['Order Date'])

data.info()
```

```python
#Converting date column
data['Order Date']=pd.to_datetime(data['Order Date'])

data.info()
```

```python
#Converting date column
data['Order Date']=pd.to_datetime(data['Order Date'])
data['Ship Date']=pd.to_datetime(data['Ship Date'])
data.info()
```

```python

data['Order Month']=data['Order Date'].dt.month
data['Order Year']=data['Order Date'].dt.year
data['Order Week']=data['Order Date'].dt.dayofweek
```

```python
#Monthly sales analysis
monthly_sales=data.groupby('Order Month')['Sales'].sum().reset_index()
monthly_sales
```

```python
fig=px.line(monthly_sales, x='Order Month', y='Sales', title='Monthly Sales Analysis')
fig.show()
```

```python
#sales by category
sales_by_category=data.groupby('Category')['Sales'].sum().reset_index()
sales_by_category
```

```python
#Donut graph
fig=px.pie(sales_by_category, values='Sales', names='Category',
           hole=0.5,color_discrete_sequence=px.colors.qualitative.Pastel)
fig.update_traces(textposition='inside', textinfo='percent+label')
fig.update_layout(title_text='Category Sales',title_font=dict(size=20))
fig.show()
```

```python
#Sub-category sales analysis
subcategory_sales=data.groupby('Sub-Category')['Sales'].sum().reset_index()
subcategory_sales
```

```python
fig=px.bar(subcategory_sales,x='Sub-Category',y='Sales',title='Sales by Sub-Category')
fig.show()
```

```python
#Monthly profit analysis
monthly_profit=data.groupby('Order Month')['Profit'].sum().reset_index()
monthly_profit
```

```python
fig=px.bar(monthly_profit, x='Order Month', y='Profit', title='Monthly Profit Analysis')
fig.show()
```

```python
#profit by category
profit_by_category=data.groupby('Category')['Profit'].sum().reset_index()
profit_by_category
```

```python
fig=px.pie(profit_by_category,
           values='Profit',
           names='Category',
           hole=0.5,
           color_discrete_sequence=px.colors.qualitative.Pastel)
fig.update_traces(textposition='inside', textinfo='percent+label')
fig.update_layout(title_text='Profit by Category', title_font=dict(size=20))
fig.show()
```

```python
#Profit by Sub-category
profit_by_subcategory=data.groupby('Sub-Category')['Profit'].sum().reset_index()
fig=px.bar(profit_by_subcategory, x='Sub-Category', y='Profit', title='Profit by Sub-Category')
fig.show()
```

```python
from plotly.graph_objs import layout
#sales and profit based on customer segment
segment_sales_profit=data.groupby('Segment').agg({'Sales':'sum', 'Profit':'sum'}).reset_index()
color_palette=colors.qualitative.Pastel
fig=go.Figure()
fig.add_trace(go.Bar(x=segment_sales_profit['Segment'],
                     y=segment_sales_profit['Sales'],name='Sales',
                     marker_color=color_palette[0]))
fig.add_trace(go.Bar(x=segment_sales_profit['Segment'],
                     y=segment_sales_profit['Profit'],name='Profit',
                     marker_color=color_palette[1]))

fig.update_layout(title='Sales and Profit by Customer Segment',
                xaxis_title='Customer Segment',
                yaxis_title='Amount')
fig.show()
```

```python
#sales by profit ratio
sales_profit_by_segment=data.groupby('Segment').agg({'Sales':'sum', 'Profit':'sum'}).reset_index()
sales_profit_by_segment['Profit Ratio']=sales_profit_by_segment['Sales']/sales_profit_by_segment['Profit']
print(sales_profit_by_segment[['Segment','Profit Ratio']])
```
